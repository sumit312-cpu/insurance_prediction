```python
##here we have to fill nulll values throuh taken avg like : mean()
#for all columns:
for i in df.columns:
  if df[i].dtype!="object":
    df[i]=df[i].fillna(df[i].mean())

```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
##steps4:  Encoding used labelencoding form convert the object columns into numeric value

for i in df.columns:
  if df[i].dtype =="object":
    df[i]=le.fit_transform(df[i])



```

```python
##steps5: detect the outliers and handle the outliers
import matplotlib.pyplot as plt
```

```python
  ## detect the outliers
for i in df.columns:
    if df[i].dtype!="object":
      plt.boxplot(df[i])
      plt.title(i)
      plt.show()
```

```python
##handle the outluers only in 2 columns we remove outlires those columns should not be categorical as well as target columns .
## if we want to keep then we can keep then no need to remove  outlires from any columns.

## colums name 1 bmi 2.Hospital_expenditure
```

```python
##HANDLE_OUTLIRES:
## for 1 column :bmi
import numpy as np
```

```python
q1=np.quantile(df['bmi'],0.25)
q1
```

```python
q3=np.quantile(df['bmi'],0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lb=q1-(1.3*(iqr))
lb
```

```python
ub=q3+(1.3*(iqr))
ub
```

```python
df=df[(df['bmi']>=lb)&(df['bmi']<=ub)]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(df["bmi"])
plt.show()
```

```python
##HANDLE_OUTLIRES:
## for 2 column :.Hospital_expenditure
import numpy as np
```

```python



q11=np.quantile(df['Hospital_expenditure'],0.25)
q1
```

```python



q31=np.quantile(df['Hospital_expenditure'],0.75)
q31
```

```python

iqr1=q31-q11
iqr1
```

```python
lb1=q11-(1.45*(iqr))
lb1
```

```python

ub1=q31+(1.45*(iqr))
ub1
```

```python


df=df[(df['Hospital_expenditure']>=lb1) & (df['Hospital_expenditure']<=ub1)]
```

```python


import matplotlib.pyplot as plt
plt.boxplot(df['Hospital_expenditure'])
plt.show()
```
